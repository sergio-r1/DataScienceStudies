# Deep Learning II


```python
import pandas as pd
from keras.models import Sequential
from keras.layers import Dense
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import confusion_matrix
from sklearn.compose import make_column_transformer
```

```python
# Carregando o banco de dados
dataset = pd.read_csv('../Dados/Credit2.csv', sep=';')
dataset
```

```python
# Serparando as variaves independentes e as variaveis dependentes
X = dataset.iloc[:,1:10].values
y = dataset.iloc[:, 10].values
```

```python
# Label encoder na coluna checking_status
labenc = LabelEncoder()
X[:,0] = labenc.fit_transform(X[:,0])

X
```

```python
# one hot encoder na coluna credit_history
oneHotEnc = make_column_transformer((OneHotEncoder(categories='auto', sparse_output=False),[1]), remainder="passthrough")
X = oneHotEnc.fit_transform(X)

X
```

```python
# Excluimos a variavel para evitar a dummy variable trap
X = (X[:,1:])

X
```

```python
# LabelEncoder com a classe
labenc_y = LabelEncoder()
y = labenc_y.fit_transform(y)

y
```

```python
# Separação do treinamento e do teste
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
print(f"xtrain {len(X_train)} xtest {len(X_test)} ytrain {len(y_train)} ytest {len(y_test)}")
```

```python
# Feature Scalling, padronização do z-score
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.fit_transform(X_test)
X_test
```

```python
classefier = Sequential()
classefier.add(Dense(units=6, kernel_initializer='uniform', activation='relu', input_dim=12))
classefier.add(Dense(units=6, kernel_initializer='uniform', activation='relu'))
classefier.add(Dense(units=1, kernel_initializer='uniform', activation='sigmoid'))
classefier.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
classefier.fit(X_train, y_train, batch_size=10, epochs=100)
```

```python
y_pred = classefier.predict(X_test)
y_pred = (y_pred>0.5)
y_pred
```

```python
cm = confusion_matrix(y_test, y_pred)
cm
```
