# Rede Neural Keras

```python
import numpy as np
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
from keras.models import Sequential
from keras.layers import Dense
from keras.utils import np_utils
```

```python
# Carregando a base de dados e criando os previsores (Variaveis independentes X) e classe (variaveis dependentes Y)
base = datasets.load_iris()
previsores = base.data # Array com as quatro colunas do arquivo
classe = base.target # Transforma valor categorico em valor numerico
classe
```

```python
# Transformação da classe para o formato dummy, pois temos uma rede neural de 3 neuronios na camada de saida
classe_dummy = np_utils.to_categorical(classe)
classe_dummy
```

```python
X_train, X_test, y_train, y_test = train_test_split(previsores, classe_dummy, test_size=0.3, random_state=0)
```

```python
# Criação da primeira estrutura da rede neural com a classe Sequential (sequencia de camadas)
modelo = Sequential() # Camadas sequencial
# primeira camada oculta 
# 5 neuronios, 4 neuronios de entrada
# Camada do tipo Dense significa que são todas conectadas
# imput_dim a respeito das 4 colunas do arquivo
modelo.add(Dense(units=5, input_dim=4)) 
# Segunda camada oculta
modelo.add(Dense(units=4))

# Terceira camada oculta
# Função de ativação = softmax
# Função softmax porque temos um problema de classificação com mais de duas classes (é gerada uma probabilidade em cada neuronio)
modelo.add(Dense(units=3, activation='softmax'))
```

```python
# Visualização da estrutura da rede neural
modelo.summary()
```

```python
# Configuração dos parametros da rede neural (adam - algoritmo para atualizar os pesos, loss o calculo do erro)
# optmizer > otimiza os pesos, loss > função de perda, metrics > para medir o erro
modelo.compile(optimizer='adam', loss='categorical_crossentropy',metrics=['accuracy'])

# Treinamento, dividindo a base de treinamento em uma porção para validação (validation_data)
modelo.fit(X_train, y_train, epochs=1000, validation_data=(X_test, y_test))
```

```python
# Previsoes e mudar a variavel para true ou false de acordo com threshold
previsoes = modelo.predict(X_test)
previsoes = (previsoes > 0.5)
len(previsoes)
```

```python
# Como é um problema com três saidas, precisamos buscar a posição que possui o maior valor (são retornados 3 valores)
y_test_matrix = [np.argmax(t) for t in y_test]
y_previsao_matrix = [np.argmax(g) for g in previsoes]
```

```python
# Geração da matriz de confusão
confusao = confusion_matrix(y_test_matrix, y_previsao_matrix)
confusao
```
